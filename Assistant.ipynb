# AI Knowledge Base Assistant — Day 1

**Goal:** Build an AI system that can answer questions about company documents such as policies, manuals, and technical guides.

Instead of manually searching through documents, users can ask natural language questions, and the system retrieves relevant information and generates answers using a Large Language Model (LLM).

**Example:**

**User:** "What is the company vacation policy?"

**AI Answer:** "Employees are entitled to 20 paid vacation days per year according to company policy."

## Project Goal

This project aims to build a system that:

- Loads company documents
- Converts them into searchable embeddings
- Stores them in a vector database
- Retrieves relevant document sections
- Uses an LLM to generate accurate answers

## Core Concepts

### 1. Large Language Models (LLMs)
A Large Language Model (LLM) is an AI model trained on massive amounts of text data.  
It can understand and generate human-like text.  

Examples of LLMs:
- GPT-4 (OpenAI)
- LLaMA (Meta)
- Gemini (Google)

LLMs can:
- Answer questions
- Summarize text
- Generate explanations
- Write code

**Limitation:** LLMs do not know private company documents, so answers may be guesses or incorrect.

---

### 2. Retrieval-Augmented Generation (RAG)
RAG combines **document search** and **LLM text generation**:



This ensures the AI produces **answers based on real documents**, not guesses.

---

### 3. LangChain
LangChain is a framework to build AI applications using LLMs.  
It simplifies:

- Document loading
- Text splitting
- Embeddings creation
- Vector database storage & search
- LLM integration

It allows developers to **quickly build RAG pipelines**.

## Project Architecture

The AI Knowledge Base Assistant works using this pipeline:

Documents 
->
Text Splitting
->
Embeddings
->
Vector Database
->
User Question
->
Similarity Search
->
LLM generates Answer

## Technologies Used

- **Python** — main programming language  
- **LangChain** — LLM pipeline framework  
- **FAISS** — vector database for embeddings  
- **OpenAI API** — LLM provider  
- **Streamlit (optional)** — web UI for chat interface
