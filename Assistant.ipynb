# AI Knowledge Base Assistant — Day 1

**Goal:** Build an AI system that can answer questions about company documents such as policies, manuals, and technical guides.

Instead of manually searching through documents, users can ask natural language questions, and the system retrieves relevant information and generates answers using a Large Language Model (LLM).

**Example:**

**User:** "What is the company vacation policy?"

**AI Answer:** "Employees are entitled to 20 paid vacation days per year according to company policy."

## Project Goal

This project aims to build a system that:

- Loads company documents
- Converts them into searchable embeddings
- Stores them in a vector database
- Retrieves relevant document sections
- Uses an LLM to generate accurate answers

## Core Concepts

### 1. Large Language Models (LLMs)
A Large Language Model (LLM) is an AI model trained on massive amounts of text data.  
It can understand and generate human-like text.  

Examples of LLMs:
- GPT-4 (OpenAI)
- LLaMA (Meta)
- Gemini (Google)

LLMs can:
- Answer questions
- Summarize text
- Generate explanations
- Write code

**Limitation:** LLMs do not know private company documents, so answers may be guesses or incorrect.

---

### 2. Retrieval-Augmented Generation (RAG)
RAG combines **document search** and **LLM text generation**:



This ensures the AI produces **answers based on real documents**, not guesses.

### 3. LangChain
LangChain is a framework to build AI applications using LLMs.  
It simplifies:

Creating PDFs for LLM Knowledge Base
Why Create PDFs?

PDFs serve as structured, shareable, and standardized documents containing information you want your LLM to understand.

They can include text, tables, and images, making them ideal for storing knowledge from multiple sources.

By converting content into PDFs, we can ensure consistency in format and ease of loading for automated processing pipelines.

PDFs allow us to test our loader with both real-world data (from the internet) and custom-written content for validation purposes.

How PDFs Are Created for This Project

Sources of PDFs:

Two PDFs from the internet: Public articles or sample documents for testing real-world data.

One self-written PDF: Custom content to validate the pipeline with controlled, known text.

Process:

Used Word to write your content.

Save or export the content as PDF.

Place the PDFs in the project folder (data/) to be read by the loader script.

Purpose of This Step

Input for the Loader Script: PDFs are the source documents that the Python script reads and converts into LLM-ready text.

Testing: Ensures the loader can handle multiple types of PDFs, including downloaded files and self-created content.

Foundation for LLM Applications: Without PDFs, there’s no structured knowledge base to query, summarize, or embed.

Ingest.py file Explanation

The code in ingest.py automatically loads all PDF files from a specified folder.

It reads each PDF page using PyPDFLoader and converts it into a document object.

All pages from all PDFs are combined into a single list of documents.

The code prints the total number of pages loaded and shows a sample of the content from the first few pages.

Key Features:

Folder scanning – Checks the data/ folder and processes all .pdf files automatically.

Page-level extraction – Each page of the PDF becomes a separate document object, making it easier to query later.

Aggregating documents – Combines all pages into one list for use in an LLM workflow.

Preview output – Prints a sample of text to verify that PDFs are loaded correctly.

Why This Code is Needed

Automation: Manually copying text from PDFs is inefficient and error-prone. This code does it automatically.

LLM-ready format: LLMs work best with structured text inputs. Converting PDFs to document objects prepares them for embedding, semantic search, and querying.

Scalability: Can handle multiple PDFs at once, including internet-sourced and self-written PDFs.

Validation: Printing a sample of loaded pages helps ensure the data is correctly ingested before further processing.

Embeddings and Vector Store

In this step, we convert text chunks into numerical representations (embeddings) and store them in a vector database for efficient semantic search.
The goal of this step is to transform text chunks into embeddings and store them in a vector database.

This enables:
- Semantic search (search by meaning)
- Fast retrieval of relevant information
- Efficient handling of large document collections

What are Embeddings?

Embeddings are numerical representations of text.

They convert words and sentences into vectors (lists of numbers) that capture their meaning.

Example:

Text:
"password policy"

Embedding:
[0.12, -0.45, 0.88, ...]

This allows the system to compare text based on meaning instead of exact words.

Why are Embeddings Important?

Without embeddings:
- Only keyword-based search is possible
- Results may be inaccurate

With embeddings:
- The system understands meaning
- Similar concepts can be matched
- Search results are more relevant

Vector Database (FAISS)

We used FAISS (Facebook AI Similarity Search) to store embeddings.

FAISS allows:
- Fast similarity search
- Efficient storage of vectors
- Scalable retrieval for large datasets

Each chunk is stored as a vector inside the database.

Observations

- Embedding generation worked efficiently using a local model
- No external API was required (cost-effective solution)
- The system is now ready for semantic search
- Vector storage enables fast retrieval in the next step

Retrieval (Semantic Search)

In this step, we implemented a retrieval system that allows users to search and find relevant information from the stored document embeddings.

The goal of this step is to retrieve relevant document chunks based on a user query.

This enables:
- Semantic search (search by meaning)
- Efficient information retrieval
- Foundation for building a chatbot

What is Retrieval?

Retrieval is the process of finding relevant information from a dataset based on a query.

Instead of matching exact keywords, semantic retrieval finds content with similar meaning.

Example:

Query:
"What is the vacation policy?"

Retrieved result:
"Employees are entitled to 20 paid vacation days per year"

How Retrieval Works

The retrieval process follows these steps:

1. Convert user query into an embedding
2. Compare it with stored embeddings in the vector database
3. Find the most similar chunks
4. Return the top relevant results

This is called similarity search.

Implementation

We used FAISS vector database along with HuggingFace embeddings.

Steps:
- Load saved FAISS vector store
- Convert user query into embedding
- Perform similarity search
- Retrieve top matching chunks

Number of results retrieved: 3 (k=3)
Results

- Successfully loaded vector database
- Retrieved top 3 most relevant chunks for each query
- System responds quickly and accurately

The retrieval system is now able to find meaningful information from documents.
Sample Query

"What is the vacation policy?"

Retrieved Output:

Employees are entitled to 20 paid vacation days per year...
 Observations

- The system retrieves relevant information even if exact keywords are not used
- Semantic search improves accuracy compared to keyword-based search
- Results are fast due to FAISS indexing

Advantages of Semantic Retrieval

- Understands meaning, not just keywords
- Works well with large document collections
- Provides more relevant results
- Scalable and efficient

Limitations

- Retrieved results are raw text (not formatted answers)
- No summarization yet
- Requires LLM integration for better responses

Chatbot (RAG - Retrieval-Augmented Generation)

In this step, we built a chatbot that can answer user questions based on the company documents using retrieval and generation techniques.
Objective

The goal of this step is to generate human-like answers using retrieved document chunks.

This enables:
- Question answering over documents
- Context-aware responses
- A more interactive AI system

What is RAG?

RAG stands for Retrieval-Augmented Generation.

It combines:
- Retrieval → finding relevant document chunks
- Generation → producing a final answer

Process:
1. User asks a question
2. System retrieves relevant chunks
3. Answer is generated using those chunks

How the Chatbot Works

The chatbot follows these steps:

1. User enters a query
2. Query is converted into an embedding
3. FAISS retrieves the most relevant chunks
4. Retrieved chunks are combined
5. A final answer is generated from the combined text

Implementation

We implemented a simple chatbot using:

- HuggingFace embeddings (local model)
- FAISS vector database
- Custom answer generation function

Steps:
- Load vector database
- Perform similarity search
- Combine top 3 results
- Display answer to the user

Results

- Chatbot successfully answers questions using document data
- Responses are generated based on retrieved chunks
- System works in real-time in the terminal

Example queries:
- What is the leave policy?
- Is remote work allowed?
- What are the security rules?

Sample Interaction

User:
"What is the vacation policy?"

Chatbot:
Based on the documents, employees are entitled to 20 paid vacation days per year...

Limitations

- Answers are not fully natural (no advanced language model)
- No summarization or reasoning beyond retrieved text
- Output depends heavily on chunk quality
